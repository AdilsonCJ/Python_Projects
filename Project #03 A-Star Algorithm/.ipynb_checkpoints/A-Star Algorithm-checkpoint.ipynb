```python
import math
import dists

# goal will always be 'bucharest'
def a_star(start, goal='Bucharest'):

    path = []

    def node_expansion(current_city):
        node_current_city = dict(dt.dists[current_city]).keys()  

        return current_city,node_current_city


    def helistic_calculation(current_city,inherited_cost=0,nodes=0):

        dict_distances = {}
        current_city,nodes = node_expansion(current_city)  

        for city in nodes:

            dict_distances[city] = None
            dict_distances[city] = int(dt.straight_line_dists_from_bucharest[city] + dict(dt.dists[current_city])[city]) 

        next_city = [k for k, it in dict_distances.items() if it == min(dict_distances.values())][0]

        if current_city == start:
            pass

        else:
            inherited_cost = dict(dt.dists[previous_city])[current_city] + inherited_cost

            for city in nodes:

                dict_distances[city] = dict_distances[city] + inherited_cost 

        return current_city,next_city,inherited_cost,dict_distances
    
    while goal not in path:
        
        if len(path) == 0:
            
            previous_city,next_city,inherited_cost = helistic_calculation(start)[0],helistic_calculation(start)[1],helistic_calculation(start)[2]
            
            path.append(previous_city)
        
        else:
            previous_city,next_city,inherited_cost = helistic_calculation(next_city,inherited_cost=inherited_cost)[0],helistic_calculation(next_city,inherited_cost=inherited_cost)[1],helistic_calculation(next_city,inherited_cost=inherited_cost)[2]
 
            path.append(previous_city)
    
        #print(path)
        
    return path
```
