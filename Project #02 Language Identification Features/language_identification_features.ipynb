<h1>Construção dos atributos</h1>
<p>Esse é o coração desse trabalho e que deverá ser desenvolvido por vocês. Pensem em como podemos "medir" cadas frase/sentença e extrair características que melhorem o resultado do processo de identificação.<p>
<p>Após a criação de cada novo atributo, execute as etapas seguintes e registre as métricas da matriz de confusão. Principalmente acurácia e a precisão.</p>

```python
import re
from unidecode import unidecode

# Evaluates the average sentence length
def tamanhoMedioFrases(texto):
    palavras = re.split("\s",texto)
    tamanhos = [len(s) for s in palavras if len(s)>0]
    soma = 0
    for t in tamanhos:
        soma=soma+t
    return soma / len(tamanhos)

# Evaluates the frequency of each letter on a phrase
def avaliar_frequencia_letras(texto_raw):

    # "Empty" - Zero Default Dictionary
    dict_alf = {'a':0,'b':0,'c':0,'d':0,'e':0,'f':0,'g':0,'h':0,'i':0,'j':0,'k':0,'l':0,'m':0,'n':0,'o':0,'p':0,'q':0,'r':0,'s':0,'t':0,'u':0,'v':0,'w':0,'x':0,'y':0,'z':0}
    
    # Text standardization 
    texto_limpo = unidecode(texto_raw.lower().replace(' ',''))

    # Count the frequency of each letter
    dicionario_contagem = {chave_letra: texto_limpo.count(chave_letra) for chave_letra in dict_alf.keys()}

    return list(dicionario_contagem.values())



def extraiCaracteristicas(frase):
  # Sentence as a vector [ 'text', 'language' ]
    texto_raw = frase[0]
    texto = frase[0]

    pattern_regex = re.compile('[^\w+]', re.UNICODE)
    texto = re.sub(pattern_regex,' ',texto)

  # Features generation
    caracteristica1 = tamanhoMedioFrases(texto)
    caracteristica2 = avaliar_frequencia_letras(texto_raw)

  # List generation
    padrao = caracteristica2
    padrao.append(caracteristica1)
    padrao.append(frase[1])
    
    return padrao


def geraPadroes(frases):
    padroes = []
    
    for frase in frases:
        padrao = extraiCaracteristicas(frase)
        padroes.append(padrao)
        
    return padroes


padroes = geraPadroes(pre_padroes)
print(padroes)

dados = pd.DataFrame(padroes)

dados
```

<h2> SVM Model Training </h1>

```python
from sklearn.model_selection import train_test_split
import numpy as np


#from sklearn.metrics import confusion_matrix

vet = np.array(padroes)
classes = vet[:,-1]         # classes = [p[-1] for p in padroes]

#print(classes)
padroes_sem_classe = vet[:,0:-1]

#print(padroes_sem_classe)
X_train, X_test, y_train, y_test = train_test_split(padroes_sem_classe, 
                                                    classes, 
                                                    test_size=0.15, 
                                                    random_state = 42,
                                                    stratify=classes)
```

```python
from sklearn import svm
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

treinador = svm.SVC()
modelo = treinador.fit(X_train, y_train)

acuracia = modelo.score(X_train, y_train)
print("Accuracy on Training Set: {:.2f}%".format(acuracia * 100))

y_pred = modelo.predict(X_train)
cm = confusion_matrix(y_train, y_pred)
print(cm)
print(classification_report(y_train, y_pred))

print('Test Score Metric')
y_pred2 = modelo.predict(X_test)
cm = confusion_matrix(y_test, y_pred2)
print(cm)
print(classification_report(y_test, y_pred2))
```
