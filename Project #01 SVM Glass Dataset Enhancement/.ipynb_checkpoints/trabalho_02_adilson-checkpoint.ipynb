```python
import pandas as pd
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns

import requests, zipfile, io
from io import BytesIO
```

## Leitura do Arquivo .data e definição das Colunas conforme Documentação
    

```python
req = requests.get('https://archive.ics.uci.edu/static/public/42/glass+identification.zip')
zpf = zipfile.ZipFile(io.BytesIO(req.content))

zpf.namelist()
dadosfp = zpf.open('glass.data')
dados = dadosfp.read()

data_glass = pd.read_csv(io.BytesIO(dados),header=None)

data_glass = data_glass.rename(columns={0:'ID',
                                        1:'RI',
                                        2:'Na',
                                        3:'Mg',
                                        4:'Al',
                                        5:'Si',
                                        6:'K',
                                        7:'Ca',
                                        8:'Ba',
                                        9:'Fe',
                                        10:'Classe',
                                       })

data_glass[['RI','Na','Mg','Al','Si','K','Ca','Ba','Fe']] = data_glass[['RI','Na','Mg','Al','Si','K','Ca','Ba','Fe']].astype(float)

data_glass['Classe'] = data_glass['Classe'].astype('category')

data_glass
```

Com o objetivo de melhorar a acurácia de 0.52 do dataset original, foi-se inicialmente validado através de um histograma que as 9 variáveis de composição dos vidros possuem grandezas muito distintas e, através de um pré-tratamento de Normalização, tivemos um aumento de quase 20 pontos percentuais no indicador (0.7192).

Contudo, pela matriz de confusão identificamos que algumas das classes tiveram 0% de precisão na predição. Ao analisar as disposições dos atributos pelas classes de interesse (ex: comparando classes 1 e 3 ou 2 / 6 e 7) verificamos que com as variáveis do dataset para o Método de SVM temos uma limitação para a separação eficiente das classes (ainda assim acima dos 50% previstos pelo UCI).

O último ponto relevante é, apesar da perda de 3 pontos percentuais na acurácia do algoritmo, caso utilizemos o Forward Propagation podemos reduzir o universo de 9 variáveis para 3 - No exemplo em questão não faz muito sentido dado o tamanho do dataset, mas é interessante ter em mente o trade-off de 3% na acurácia pela eliminação de 66% dos atributos "desnecessários".
